```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.style.use('ggplot')
```

```python
suicides = pd.read_csv(r'E:\Projects\data-analysis-examples\master.csv')
```

```python
suicides
```

```python
suicides[['country', 'age', 'suicides_no']]
```

```python
suicides[(suicides['sex'] == 'male') & (suicides['country'] == 'Czech Republic')]
```

```python
suicides.groupby('sex')['suicides_no'].sum().plot.pie(title='poměr sebevražd')
plt.savefig('poměr sebevražd.png')
```

```python
suicides.groupby('sex')['suicides_no'].sum().plot.bar(title='poměr sebevražd - barchart')
plt.savefig('poměr sebevražd - barchart.png')
```

```python
ages_order = ['5-14 years', '15-24 years', '25-34 years', '35-54 years', '55-74 years', '75+ years']
suicides.groupby('age')['suicides_no'].sum().reindex(ages_order).plot.bar()
plt.savefig('sebevrazdy-vek.png')
```

```python
suicides.groupby('age')['suicides_no'].sum().reindex(ages_order).plot.pie(figsize=(10, 10))
```

```python
suicides[suicides['country'] == 'Czech Republic'].groupby('age')['suicides_no'].sum()
```

```python
suicides_czech = suicides[suicides['country'] == 'Czech Republic']
suicides_czech.groupby('year')['suicides_no'].sum().plot()
```

```python
czech_sex_suicides = suicides_czech.groupby(['year', 'sex'])['suicides_no'].sum().reset_index()
czech_sex_suicides
```

```python
czech_male_suicides = czech_sex_suicides[czech_sex_suicides['sex'] == 'male']
czech_female_suicides = czech_sex_suicides[czech_sex_suicides['sex'] == 'female']
czech_male_suicides.plot(x='year', y='suicides_no', label='muži', ax=plt.gca())
czech_female_suicides.plot(x='year', y='suicides_no', label='ženy', ax=plt.gca())
plt.savefig('sebevraždy česko v čase.png')
```

```python
suicides_country_sex_year = suicides.groupby(['country', 'sex', 'year']).sum().reset_index()
suicides_country_sex_year
```

```python
countries = ['Czech Republic', 'Spain', 'Germany', 'France']
my_suicides = suicides_country_sex_year[suicides_country_sex_year['country'].isin(countries)]
sns.lineplot(x='year', y='suicides_no', hue='country', data=my_suicides)
```

```python
countries = ['Czech Republic', 'Spain', 'Germany', 'France', 'Uzbekistan']
my_suicides = suicides_country_sex_year[suicides_country_sex_year['country'].isin(countries)]
my_suicides = my_suicides.groupby(['country', 'year']).sum().reset_index()
```

```python
plt.figure(figsize=(12, 8))
sns.lineplot(x='year', y='suicides_no', hue='country', data=my_suicides)
```

```python
plt.figure(figsize=(12, 8))
sns.lineplot(x='year', y='suicides/100k pop', hue='country', data=my_suicides)
```

```python
plt.figure(figsize=(12, 8))
sns.lineplot(x='year', y='suicides/100k pop', hue='country', data=my_suicides[my_suicides['country'] == 'Uzbekistan'])
```

```python
plt.figure(figsize=(12, 8))
sns.regplot(x='year', y='suicides/100k pop', data=my_suicides[my_suicides['country'] == 'Uzbekistan'])
```
