```python
from statsmodels.tsa.arima_model import ARIMA
```

```python
czech_suicides_time = suicides_czech.groupby('year')['suicides_no'].sum()
czech_suicides_time
```

```python
model = ARIMA(czech_suicides_time, order=(2,1,0))
model_fit = model.fit()
suicides_predicted, suicides_err, suicides_conf_intv = model_fit.forecast(3)
suicides_predicted_series = pd.Series(suicides_predicted, index=(2017, 2018, 2019))
suicides_predicted_series
```

```python
suicides_with_prediction = czech_suicides_time.append(suicides_predicted_series)
suicides_with_prediction
```

```python
czech_suicides_time.plot(style='r-')
suicides_predicted_series.plot(style='r.')
```

```python
czech_suicides_years = czech_suicides_time.to_frame().reset_index()
sns.regplot(x='year', y='suicides_no', data=czech_suicides_years)
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
import numpy as np

poly = PolynomialFeatures(degree=20)
czech_suicides_years_poly = poly.fit_transform(np.array(czech_suicides_years['year'])[:, np.newaxis])
czech_suicides_years_poly
```

```python
lin = LinearRegression()
lin.fit(czech_suicides_years_poly, czech_suicides_years['suicides_no'])
lin.predict(poly.transform(np.array([[2018], [2019], [2020]])))
```

```python
pd.DataFrame(czech_suicides_years_poly)
```

```python
years_to_see = np.concatenate((czech_suicides_years['year'].to_numpy(), np.array([2017, 2018, 2019, 2020])))
years_to_see
```

```python
fitted_suicides = lin.predict(poly.transform(years_to_see[:, np.newaxis]))
fitted_suicides
```

```python
fitted_suicides_pd = pd.Series(fitted_suicides, index=years_to_see, name='suicides_no')
fitted_suicides_pd
```

```python
plt.figure()
fitted_suicides_pd.plot(ax=plt.gca())
czech_suicides_years.plot.scatter(x='year', y='suicides_no',ax=plt.gca())
```
